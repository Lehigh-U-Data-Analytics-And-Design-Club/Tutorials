# Plotly and Cufflinks

Plotly is a library that allows you to create interactive plots that you can use in dashboards or websites (you can save them 
as html files or static images).  
See https://plot.ly

Cufflinks binds the power of plotly with the flexibility of pandas for easy plotting.  
See https://github.com/santosjorge/cufflinks




## Installation

In order for this all to work, you'll need to install plotly and cufflinks to call plots directly off of a pandas dataframe. These libraries are not currently available through **conda** but are available through **pip**. Install the libraries at your command line/terminal using:

    pip install plotly
    pip install cufflinks


## Imports and Set-up

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
import cufflinks as cf
```

```python
# For Notebooks
init_notebook_mode(connected=True)
```

```python
# For offline use
cf.go_offline()
```

### Create some data

```python
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})
```

```python
df2.head()
```

## Using Cufflinks and iplot()

* scatter
* bar
* box
* spread
* ratio
* heatmap
* surface
* histogram
* bubble

```python
df.plot()
```

```python
df.iplot()
```

## Scatter

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers',size=10)
```

## Bar Plots

```python
df2.iplot(kind='bar',x='Category',y='Values')
```

```python
df.head()
```

```python
df.iplot(kind='bar')
```

```python
df.count().iplot(kind='bar')
```

```python
df.sum().iplot(kind='bar')
```

## Boxplots

```python
df.iplot(kind='box')
```
